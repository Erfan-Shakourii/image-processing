# Convex Hull

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
np.random.seed=50
random_points = np.random.randint(100, 400, 40)
random_points = random_points.reshape(20, 2)
image = np.zeros((512,512,3), np.uint8)
for point in random_points:
    cv2.circle(image, (point[0],point[1]),8,(0,250,250),-1)
plt.imshow(image[...,::-1])
```

```python
hull = cv2.convexHull(random_points)
cv2.drawContours(image, [hull], -1, (255,0,0), 4)
plt.imshow(image[...,::-1])
```

```python
bgr_image = cv2.imread("images/star.png")

src_gray = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2GRAY)

ret, thresh = cv2.threshold(src_gray, 127, 255, 0)

contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

hull_list = []
for i in range(len(contours)):
    hull = cv2.convexHull(contours[i])
    hull_list.append(hull)
    
cv2.drawContours(bgr_image, contours, -1, (255,255,0), 1)
cv2.drawContours(bgr_image, hull_list, -1, (0,255,255),1)

plt.figure(figsize=[7,7])
plt.imshow(bgr_image[...,::-1])
```

```python
bgr_image = cv2.imread("images/shapes.png")

src_gray = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2GRAY)

ret, thresh = cv2.threshold(src_gray, 127, 255, 0)

contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

hull_list = []
for i in range(len(contours)):
    hull = cv2.convexHull(contours[i])
    hull_list.append(hull)
    
cv2.drawContours(bgr_image, contours, -1, (255,255,0), 2)
cv2.drawContours(bgr_image, hull_list, -1, (0,255,255),1)

plt.figure(figsize=[7,7])
plt.imshow(bgr_image[...,::-1])
```
